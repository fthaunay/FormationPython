```python
import pandas as pd
```

```python
twitter_data = pd.read_csv("twitter-DataBzh-2016-12.csv")
```

```python
twitter_data.head()
```

```python
from datetime import datetime
from dateutil import parser
```

Alternative a datetime, dateutil facilite la mise en forme sans devoir introcuire le format
Obligatoire pour traiter les timezone

```python
twitter_data.heure = [parser.parse(str(date), ignoretz=True) for date in twitter_data.heure]
```

```python
twitter_data.head()
```

```python
# unicodedata pour le formatage de texte
import unicodedata
```

```python
# comment retirer les  accents et cedilles des textes
# penser a convertir la string en unicode avant de pouvoir l'encoder en ASCII
twitter_data["Texte du Tweet"].head()
twitter_data["Texte du Tweet"]=[unicodedata.normalize("NFKD", unicode(x, "utf-8"))
                                .encode("ASCII", "ignore") for x in twitter_data["Texte du Tweet"]]
```

```python
twitter_data.head()
```

```python
(twitter_data.index)
```

```python
twitter_data.columns=[unicodedata.normalize("NFKD", unicode(x, "utf-8"))
                                .encode("ASCII", "ignore") for x in twitter_data.columns]
```
